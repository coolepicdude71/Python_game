```
import pygame
import random

# Initialize pygame
pygame.init()

# Game window
width, height = 600, 400
cell = 20

screen = pygame.display.set_mode((width, height))
pygame.display.set_caption("Snake Game")

clock = pygame.time.Clock()

# Snake
snake = [(100, 100), (80, 100), (60, 100)]
direction = (20, 0)

# Food
food = (random.randrange(0, width, cell), random.randrange(0, height, cell))

running = True

while running:

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    keys = pygame.key.get_pressed()

    if keys[pygame.K_UP]:
        direction = (0, -20)
    if keys[pygame.K_DOWN]:
        direction = (0, 20)
    if keys[pygame.K_LEFT]:
        direction = (-20, 0)
    if keys[pygame.K_RIGHT]:
        direction = (20, 0)

    # Move snake
    head = (snake[0][0] + direction[0], snake[0][1] + direction[1])
    snake.insert(0, head)

    if head == food:
        food = (random.randrange(0, width, cell), random.randrange(0, height, cell))
    else:
        snake.pop()

    # Collision
    if head in snake[1:] or head[0] < 0 or head[1] < 0 or head[0] >= width or head[1] >= height:
        running = False

    # Draw
    screen.fill((0,0,0))

    for segment in snake:
        pygame.draw.rect(screen, (0,255,0), (*segment, cell, cell))

    pygame.draw.rect(screen, (255,0,0), (*food, cell, cell))

    pygame.display.update()
    clock.tick(10)

pygame.quit()
```
